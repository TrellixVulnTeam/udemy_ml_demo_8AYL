```python
import matplotlib.pyplot as plt
import numpy as np

a = np.linspace(0,10,11)
b = a ** 4
x = np.arange(0,10)
y = 2 * x
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[1].plot(a,b)
```
